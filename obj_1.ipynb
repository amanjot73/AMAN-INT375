```python
# this is for  Temperature Trends
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel('my new datsset.xlsx')
# Create datetime column
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
# Temperature Analysis Code
def analyze_temperature_trends():
    plt.figure(figsize=(12, 6))
    
    # Create line plot for each city
    for city in df['city'].unique():
        city_data = df[df['city'] == city]
        plt.plot(city_data['date'], 
                city_data['temperature_celsius'], 
                label=city)
    
    plt.title('Temperature Trends Over Time by City', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Temperature (C)', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # Monthly temperature distribution
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x=df['date'].dt.month, y='temperature_celsius')
    plt.title('Monthly Temperature Distribution')
    plt.xlabel('Month')
    plt.ylabel('Temperature (C)')
    plt.show()

analyze_temperature_trends()
```

```python
# this is for  Humidity Levels
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel('my new datsset.xlsx')
# Create datetime column
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
# Humidity Analysis Code
def analyze_humidity_levels():
    plt.figure(figsize=(12, 6))
    
    # Humidity trends over time
    sns.lineplot(data=df, x='date', 
                y='relative_humidity_percent', 
                hue='city')
    
    plt.title('Humidity Trends Over Time', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Relative Humidity (%)', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Correlation between humidity and temperature
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, 
                    x='temperature_celsius', 
                    y='relative_humidity_percent', 
                    hue='city')
    plt.title('Temperature vs Humidity Correlation')
    plt.xlabel('Temperature (C)')
    plt.ylabel('Relative Humidity (%)')
    plt.show()

analyze_humidity_levels()
```

```python
# this is for  Wind Speed Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel('my new datsset.xlsx')
# Create datetime column
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
# Wind Speed Analysis Code
def analyze_wind_patterns():
    plt.figure(figsize=(12, 6))
    
    # Wind speed trends
    sns.lineplot(data=df, x='date', 
                y='wind_speed_10m', 
                hue='city')
    
    plt.title('Wind Speed Trends Over Time', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Wind Speed (m/s)', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Wind speed distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x='wind_speed_10m', 
                hue='city', multiple="stack")
    plt.title('Wind Speed Distribution by City')
    plt.xlabel('Wind Speed (m/s)')
    plt.ylabel('Count')
    plt.show()

analyze_wind_patterns()
```

```python
# this is for Rain Forecast Tracking
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel('my new datsset.xlsx')
# Create datetime column
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
# Rain Analysis Code
def analyze_rainfall():
    plt.figure(figsize=(12, 6))
    
    # Daily precipitation
    sns.lineplot(data=df, x='date', 
                y='precipitation', 
                hue='city')
    
    plt.title('Precipitation Over Time', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Precipitation (mm)', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Monthly rainfall patterns
    monthly_rain = df.groupby(['city', df['date'].dt.month])['precipitation'].mean().reset_index()
    
    plt.figure(figsize=(10, 6))
    sns.barplot(data=monthly_rain, x='date', 
                y='precipitation', hue='city')
    plt.title('Average Monthly Precipitation by City')
    plt.xlabel('Month')
    plt.ylabel('Average Precipitation (mm)')
    plt.show()

analyze_rainfall()

```

```python
# this is for Extreme Weather Alerts
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel('my new datsset.xlsx')
# Create datetime column
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
# Extreme Weather Analysis Code
def analyze_extreme_weather():
    # Define thresholds for extreme conditions
    TEMP_THRESHOLD = 35  # C
    WIND_THRESHOLD = 20  # m/s
    RAIN_THRESHOLD = 50  # mm
    
    # Create extreme weather indicators
    df['extreme_temp'] = df['temperature_celsius'] > TEMP_THRESHOLD
    df['extreme_wind'] = df['wind_speed_10m'] > WIND_THRESHOLD
    df['extreme_rain'] = df['precipitation'] > RAIN_THRESHOLD
    
    # Plot extreme conditions
    fig, axes = plt.subplots(3, 1, figsize=(12, 15))
    
    # Temperature extremes
    sns.scatterplot(data=df, x='date', 
                    y='temperature_celsius', 
                    hue='extreme_temp', 
                    ax=axes[0])
    axes[0].set_title('Extreme Temperature Events')
    axes[0].set_xlabel('Date')
    axes[0].set_ylabel('Temperature (C)')
    
    # Wind extremes
    sns.scatterplot(data=df, x='date', 
                    y='wind_speed_10m', 
                    hue='extreme_wind', 
                    ax=axes[1])
    axes[1].set_title('Extreme Wind Events')
    axes[1].set_xlabel('Date')
    axes[1].set_ylabel('Wind Speed (m/s)')
    
    # Rain extremes
    sns.scatterplot(data=df, x='date', 
                    y='precipitation', 
                    hue='extreme_rain', 
                    ax=axes[2])
    axes[2].set_title('Extreme Rainfall Events')
    axes[2].set_xlabel('Date')
    axes[2].set_ylabel('Precipitation (mm)')
    
    plt.tight_layout()
    plt.show()

analyze_extreme_weather()
```
